```python
import torch
import matplotlib.pyplot as plt
from deeprecon.torch.projector import projector3d
from deeprecon.torch.utils.phantom import phantom
```

```python
x = phantom(n=(256, 256, 10), dim=3)
x = torch.from_numpy(x).to(torch.float32).to("cuda")
x = x.permute(2, 0, 1)
x = x[None, None, ...]
projector = projector3d((256, 256, 10), (256, 10), 1, 2, 500, 1000, scan_type="flat")
projector.to("cuda")
views = torch.arange(256) * torch.pi * 2 / 256
views = views.to("cuda")
```

```python
p = projector.projection(x, views)
b = projector.backprojection(p, views)
recon = projector.filtered_backprojection(p, views)
```

```python
plt.figure()
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x[:, :, i].cpu().numpy().squeeze(), cmap="gray")
plt.show()
```

```python
plt.figure()
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(p[..., i, :].cpu().numpy().squeeze(), cmap="gray")
plt.show()
```

```python
plt.figure()
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(b[:, :, i].cpu().numpy().squeeze(), cmap="gray")
plt.show()
```

```python
plt.figure()
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(recon[:, :, i].cpu().numpy().squeeze(), cmap="gray")
plt.show()
```
